# Chi-Square Distribution as sum of squared Standard Normal Random Variables

The $\chi^2$ distribution is the sum of square of Standard Normal random variables.

If $x_1, x_2,...,x_n$ are $n$ independent $N[0,1]$ random variables, then $\sum_{i=1}^{n} x_i^2 \sim \chi^2[n]$.

I want to test this graphically by simulation.

```python
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
from scipy.stats import chi2
```

```python
## First let us draw some standard normal random variables

x1 = np.random.normal(0,1,1000000)
x2 = np.random.normal(0,1,1000000)
x3 = np.random.normal(0,1,1000000)
x4 = np.random.normal(0,1,1000000)
x5 = np.random.normal(0,1,1000000)
x6 = np.random.normal(0,1,1000000)
x7 = np.random.normal(0,1,1000000)
x8 = np.random.normal(0,1,1000000)
x9 = np.random.normal(0,1,1000000)

x1_squared = x1**2
x2_squared = x2**2
x3_squared = x3**2
x4_squared = x4**2
x5_squared = x5**2
x6_squared = x6**2
x7_squared = x7**2
x8_squared = x8**2
x9_squared = x9**2

chi1 = x1_squared
chi_2 = x1_squared + x2_squared
chi3 = x1_squared + x2_squared + x3_squared
chi4 = x1_squared + x2_squared + x3_squared + x4_squared
chi6 = x1_squared + x2_squared + x3_squared + x4_squared + x5_squared + x6_squared
chi9 = x1_squared + x2_squared + x3_squared + x4_squared + x5_squared + x6_squared + x7_squared + x8_squared + x9_squared
```

```python
x = np.arange(0, 20, 0.001)

fig, ax = plt.subplots()
ax.set_ylim(top=0.3)
ax.plot(x, chi2.pdf(x, df=1), color="blue", label="Dof = 1")
ax.plot(x, chi2.pdf(x, df=2), color="red", label="Dof = 2")
ax.plot(x, chi2.pdf(x, df=3), color="green", label="Dof = 3")
ax.plot(x, chi2.pdf(x, df=4), color="violet", label="Dof = 4")
#ax.hist(chi4, bins=40, facecolor="violet",density="True", edgecolor="white", linewidth=2, alpha=0.3)
ax.plot(x, chi2.pdf(x, df=6), color="purple", label="Dof = 6")
ax.plot(x, chi2.pdf(x, df=9), color="yellow", label="Dof = 9")
ax.legend(loc="upper right")
#ax.hist(chi9, bins=40, facecolor="blue", density="True", edgecolor="white", linewidth=2, alpha=0.3)
plt.show()
```
